## Start Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../data/Housing.csv")
df_temp = df.copy()
df_temp.info()
```

## BOXPLOT

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_temp['price'])
plt.title("Boxplot de price pour d√©tecter les outliers")
plt.xlabel("Price")
plt.show()
```

## Scatter Plot

```python
# Exemple de Scatter Plot pour visualiser 'LoanAmount' par rapport au 'ApplicantIncome'
plt.figure(figsize=(10, 6))
plt.scatter(df_temp['price'], df_temp['area'], alpha=0.5)
plt.title("Scatter Plot de price vs area")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

## Histogramme

```python
# Exemple d'Histogramme pour visualiser la distribution de 'LoanAmount'
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=30, edgecolor='black')
plt.title("Histogramme de price")
plt.xlabel("price")
plt.ylabel("area")
plt.show()
```

## Pairplot

```python
# Exemple de Pairplot pour voir les relations entre plusieurs variables
sns.pairplot(df[['price', 'area', 'bedrooms', 'bathrooms', 'stories']])
plt.show()
```
